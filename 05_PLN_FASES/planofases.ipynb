```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def sistema_ecuaciones(y, t):
    dy1_dt = -y[1] + y[0]*(1 - y[0]**2 - y[1]**2)
    dy2_dt = y[0] + y[1]*(1 - y[0]**2 - y[1]**2)
    
    return [dy1_dt, dy2_dt]

# Condiciones iniciales
y0 = [0.5, 0.5]

# Tiempo de integración
t = np.linspace(0, 20, 1000)

# Solución del sistema de ecuaciones usando odeint
sol = odeint(sistema_ecuaciones, y0, t)

# Ploteo del plano de fases
plt.plot(sol[:, 0], sol[:, 1])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plano de Fases del Sistema de Ecuaciones')
plt.grid(True)
plt.show()
```
